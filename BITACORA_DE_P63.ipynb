# **NOTA IMPORTANTE PARA LA CALIFICACIÓN: Antes de ejecutar la siguiente  celda, es estrictamente necesario descargar los 5 archivos FASTA del enlace de Google Drive de arriba y subirlos manualmente al panel de archivos izquierdo de Google Colab.**

```python
%%bash
# 1. Unificar las secuencias seleccionadas en un archivo multi-FASTA
cat p63_protein.fasta sequence_RANA.fasta PEZ_MEXICANO.fasta lagartodPared.fasta sequence_patito.fasta > homologos.fasta

# 2. Limpiar las descripciones secundarias de las cabeceras con expresiones regulares
sed -i 's/\s.*//' homologos.fasta

# 3. Ejecutar el alineamiento global con la sintaxis de MUSCLE v5
./muscle -align homologos.fasta -output alineamiento_ofp63.txt
```
